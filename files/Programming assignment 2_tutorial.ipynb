```python
import pandas as pd
```

```python
data = pd.read_csv("../PRML TA/iris.csv")
iris=np.array(data)
```

```python
def cal_mean(data):
    mean = np.mean(data)
    return mean
def cal_var(data, mean,n):
    var = np.dot((data-mean),(data-mean)) /(n-1)
    return var
# mean of each feature each class
setosa_mean_f1 = cal_mean(iris[0:35,2])
vesicolor_mean_f1 = cal_mean(iris[50:85,2])
verginica_mean_f1 = cal_mean(iris[100:135,2])
setosa_mean_f2 = cal_mean(iris[0:35,3])
vesicolor_mean_f2 = cal_mean(iris[50:85,3])
verginica_mean_f2 = cal_mean(iris[100:135,3])

# var of each feature each class
setosa_var_f1 = cal_var(iris[0:35,2],setosa_mean_f1,50)
vesicolor_var_f1 = cal_var(iris[50:85,2],vesicolor_mean_f1,35)
verginica_var_f1 = cal_var(iris[100:135,2],verginica_mean_f1,35)
setosa_var_f2 = cal_var(iris[0:35,3],setosa_mean_f2,35)
vesicolor_var_f2 = cal_var(iris[50:85,3],vesicolor_mean_f2,35)
verginica_var_f2 = cal_var(iris[100:135,3],verginica_mean_f2,35)
```

```python
test_data = np.vstack((iris[35:50],iris[85:100],iris[135:150]))
```

```python
test_data.shape
```

```python
class_decisons_list=[]
actual_class_list=[]
for i in range(test_data.shape[0]):
    p_setosa = -1/2 * (2*np.log(2*np.pi)+np.log(setosa_var_f1)+np.log(setosa_var_f2)\
                       +((test_data[i,2]-setosa_mean_f1)**2)*1/setosa_var_f1\
                       +((test_data[i,3]-setosa_mean_f2)**2)*1/setosa_var_f2)
    
    #try to do this calculations using a funtion that is easier
    
    p_vesicolor = -1/2 * (2*np.log(2*np.pi)+np.log(vesicolor_var_f1)+np.log(vesicolor_var_f2)\
                       +((test_data[i,2]-vesicolor_mean_f1)**2)*1/vesicolor_var_f1\
                       +((test_data[i,3]-vesicolor_mean_f2)**2)*1/vesicolor_var_f2)
    
    p_verginica = -1/2 * (2*np.log(2*np.pi)+np.log(verginica_var_f1)+np.log(verginica_var_f2)\
                       +((test_data[i,2]-verginica_mean_f1)**2)*1/verginica_var_f1\
                       +((test_data[i,3]-verginica_mean_f2)**2)*1/verginica_var_f2)
    compare_lkh = [p_setosa,p_vesicolor,p_verginica]
    class_decisons_list.append(compare_lkh.index(max(compare_lkh)))
    if test_data[i,5]=='Iris-setosa':
        actual_class_list.append(0)
    elif test_data[i,5]=='Iris-versicolor':
        actual_class_list.append(1)
    elif test_data[i,5]=='Iris-virginica':
        actual_class_list.append(2)
    
```

```python
accuracy_list = list(i[0] == i[1] for i in zip(class_decisons_list, actual_class_list))
A = np.array(accuracy_list,dtype=int)
no_of_corrects = np.dot(A,A)
accuracy_rate = (no_of_corrects/len(accuracy_list))*100
print("the Accuracy rate is ", accuracy_rate,"%")
```
